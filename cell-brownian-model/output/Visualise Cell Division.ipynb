```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
import imageio
import random
from mpl_toolkits.mplot3d import Axes3D
```

```python
def print_a_circle(positions, r, my_color):    
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)

    x_centre = positions[0]        
    y_centre = positions[1]        
    z_centre = positions[2]    
    x = r * np.outer(np.cos(u), np.sin(v)) + x_centre
    y = r * np.outer(np.sin(u), np.sin(v)) + y_centre
    z = r * np.outer(np.ones(np.size(u)), np.cos(v)) + z_centre
    ax.plot_surface(x, y, z,  rstride=4, cstride=4, color=my_color, linewidth=0, alpha=0.3)
```

```python
def swap_to_two_circles(positions, r):
    theta = random.uniform(0, np.pi)
    phi = random.uniform(0, 2*np.pi)
    position1 = [positions[0] + 0.5*r*np.sin(theta)*np.cos(phi), positions[1] + 0.5*r*np.sin(theta)*np.sin(phi),
                     positions[2] + 0.5*r*np.cos(phi)]
    position2 = [positions[0] - 0.5*r*np.sin(theta)*np.cos(phi), positions[1] - 0.5*r*np.sin(theta)*np.sin(phi),
                     positions[2] - 0.5*r*np.cos(phi)]
    return position1, position2
    
    
```

```python
initial_positions = []
for i in range (0,3):
    initial_positions.append(random.uniform(10,100))
print(initial_positions)
end_position1, end_position2 = swap_to_two_circles(initial_positions, 0.5)
```

```python
fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
ax = Axes3D(fig)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
for ii in range(0,360,100):
    ax.view_init(elev=10., azim=ii)
    print_a_circle(initial_positions, 0.5, "r")
    print_a_circle(end_position1, 0.5, "b")
    print_a_circle(end_position2, 0.5, "b")
    plt.savefig("azim={}".format(ii))
#plt.close()

```
