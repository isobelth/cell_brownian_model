```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_lines = []
with open ("output.txt") as my_file:
    for line in my_file:
        line = line.strip()
        temp = line.split(",")
        file_lines.append([float(temp[0]), float(temp[1])])       
```

```python
x_max = (np.array(file_lines).max(axis=0))[0]
y_max = (np.array(file_lines).max(axis=0))[1]
x_min = (np.array(file_lines).min(axis=0))[0]
y_min = (np.array(file_lines).min(axis=0))[1]
```

```python
def print_the_output(file_lines, start_row, end_row, color_to_plot):
    circles = []
    for i in range (start_row, end_row):
        circle = plt.Circle((file_lines[i][0], file_lines[i][1]), 0.05, fill = True, color = color_to_plot)
        circles.append(circle)
    return circles
    
```

```python
max_line = (len(file_lines))
start_line = 0
while start_line < max_line - 10:
    #ax.set_title("Start line=", start_line)
    plt.figure(figsize=(7,7))
    ax = plt.gca()
    ax.set_xlim((x_min, x_max))
    ax.set_ylim((y_min, y_max))
    ax.set_aspect('equal')
    for circle in (print_the_output(file_lines, start_line, start_line + 10, 'r')):
        ax.add_artist(circle)
    plt.savefig("start line = {}".format(start_line))
    
    start_line += 10
```
